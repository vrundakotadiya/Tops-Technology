30) How will you compare two lists? 

You can compare two lists by:

✔ Exact match → list1 == list2
if list1 == list2:
    print("Both lists are equal.")
else:
    print("Lists are not equal.")

✔ Same elements ignoring order → sorted(list1) == sorted(list2)
if sorted(list1) == sorted(list2):
    print("Lists have the same elements.")
else:
    print("Lists are different.")

✔ Differences → set(list1) ^ set(list2)
diff = set(list1) ^ set(list2)
print("Different elements:", diff)

✔ Common elements → set(list1) & set(list2)
common = set(list1) & set(list2)

43) What is tuple? Difference between list and tuple. 

Ans:- A tuple is a collection of items in Python that is ordered and unchangeable (immutable).
It is written with round brackets ( ).

| Feature               | **List**                                 | **Tuple**                         |
| --------------------- | ---------------------------------------- | --------------------------------- |
| **Syntax**            | Written with square brackets `[ ]`       | Written with round brackets `( )` |
| **Mutability**        | Mutable (can be changed)                 | Immutable (cannot be changed)     |
| **Speed**             | Slower                                   | Faster                            |
| **Memory usage**      | Uses more memory                         | Uses less memory                  |
| **Methods available** | Many (like `append()`, `remove()`, etc.) | Few (like `count()`, `index()`)   |
| **Use case**          | When data can change                     | When data should stay constant    |

```python
# 48)Write a Python script to sort (ascending and descending) a dictionary by value. 

mydict = {'a': 45, 
          'b': 12,
          'c': 89, 
          'd': 33
        }
ascending=dict(sorted(mydict.items(),key=lambda x:x[1]))
print("Ascending:",ascending)

descending=dict(sorted(mydict.items(), key=lambda x:x[1],reverse=True))
print("Descending:",descending)
```

51) How Do You Traverse Through a Dictionary Object in Python?

Ans:- You can traverse (iterate) through a dictionary in Python using loops. The common ways are:

1️)Traverse through keys
for key in my_dict:
    print(key)

2️)Traverse through values
for value in my_dict.values():
    print(value)

3)Traverse through key–value pairs
for key, value in my_dict.items():
    print(key, value)

52) How Do You Check the Presence of a Key in A Dictionary?

Ans:- You can check if a key exists in a dictionary using the in operator,
get() method, or keys() method.

1️)Using in operator (Most common & recommended)
if 'name' in my_dict:
    print("Key exists")

2️)Using get() method
if my_dict.get('name') is not None:
    print("Key exists")

3️)Using keys() method
if 'name' in my_dict.keys():
    print("Key exists")
